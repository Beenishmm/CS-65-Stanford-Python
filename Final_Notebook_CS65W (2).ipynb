# **Data Visualization**
In summary, this function generates a visually appealing and informative line chart that illustrates the cumulative frequency of songs released by the top artists on Spotify over the years. It includes various styling elements to enhance readability and professional appearance.

```python
import matplotlib.pyplot as plt

def plot_customized_line_chart(dataframe, top_n=5):
    # Get the top artists and their frequencies
    top_artists = dataframe['artist(s)_name'].value_counts().head(top_n)

    # Define a custom color palette for the lines with better contrast
    colors = ['#e3120b', '#1e90ff', '#ff4500', '#008000', '#ffa500']

    # Create a figure with a light gray background
    fig, ax = plt.subplots(figsize=(8, 6), dpi=100, facecolor='#f0f0f0')

    # Plot lines for each artist with markers
    for artist, color in zip(top_artists.index, colors):
        artist_data = dataframe[dataframe['artist(s)_name'] == artist]
        artist_data = artist_data.groupby('released_year').size().cumsum()
        plt.plot(artist_data.index, artist_data.values, label=artist, color=color, linewidth=2, marker='o', markersize=5)

    # Add labels and title
    plt.title('Top Artists on Spotify', fontsize=14, fontweight='bold')
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Cumulative Frequency', fontsize=12)

    # Set a legend with a more professional appearance
    legend = plt.legend(title='Artist', fontsize=10)
    legend.get_title().set_fontsize(12)
    legend.get_title().set_weight('bold')

    # Remove the top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Set y-axis major ticks at specified intervals
    plt.yticks(range(0, max(artist_data) + 1, 10), fontsize=10)

    # Add data labels for key points on the lines
    for artist, color in zip(top_artists.index, colors):
        artist_data = dataframe[dataframe['artist(s)_name'] == artist]
        artist_data = artist_data.groupby('released_year').size().cumsum()
        # Add a label to the last data point
        plt.text(artist_data.index[-1], artist_data.values[-1], artist, fontsize=10, color=color, va='bottom')

    # Adjust padding and spacing for a cleaner appearance
    plt.tight_layout()

    # Add source attribution
    plt.figtext(0.15, 0.01, 'Source: Spotify Data', fontsize=8, color='gray')

    # Remove grid lines from behind
    ax.yaxis.grid(False)

    # Save the plot as an image (e.g., PNG)
    plt.savefig('customized_line_chart.png', bbox_inches='tight')

    # Show the plot
    plt.show()


plot_customized_line_chart(data1)
```

# In summary, this code segment loads a Spotify dataset, cleans the specified column by converting it to numeric values, and then creates and visualizes a histogram of the data. Outliers are identified and highlighted in red on the histogram for better insight into the distribution of the data.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your Spotify dataset
df = pd.read_csv('spotify-2023.csv', encoding = 'latin-1')

# Specify the column for which you want to create a histogram
column_name = 'in_shazam_charts'

# Convert the column to numeric, handling errors by coercing to NaN
df[column_name] = pd.to_numeric(df[column_name], errors='coerce')

# Plot the histogram
plt.figure(figsize=(10, 6))
plt.hist(df[column_name].dropna(), bins=20, color='skyblue', edgecolor='black', alpha=0.7)

# Highlight outliers in red
outliers = df[df[column_name] > df[column_name].mean() + 2 * df[column_name].std()]
plt.scatter(outliers[column_name], [0] * len(outliers), color='red', marker='o', label='Outliers')

# Set labels and title
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name} with Outliers Highlighted')
plt.legend()

# Show the plot
plt.show()
```

# **#Making elbow curve**

# **The Elbow Method is used to determine the optimal number of clusters for K-Means clustering. The plot shows how the WCSS decreases as the number of clusters increases. The "elbow" point, where the rate of decrease sharply changes, is typically considered the optimal number of clusters. This information helps in deciding the appropriate number of clusters to use in further analysis.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Assuming data2 is already defined
X = data2[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean']]

# Standardize the features
X_scaled = StandardScaler().fit_transform(X)

# Elbow method to find the optimal number of clusters
wcss = [KMeans(n_clusters=i, random_state=42).fit(X_scaled).inertia_ for i in range(1, 11)]

# Plot the elbow curve with a smaller figure size
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()


```

# PCA
In summary, this code performs Principal Component Analysis (PCA) on a subset of features from the data2 DataFrame, standardizes the features, transforms them into three principal components, and then visualizes the results with scatter plots of different combinations of principal components. The scatter plots provide insights into the distribution and relationships among the data points in the reduced-dimensional space.

```python
from sklearn.decomposition import PCA
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Select the features for PCA
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']

# Extract features and target variable
X = data2[features]
y = data2['diagnosis']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA with two components for visualization
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)

# Create a DataFrame with the principal components and the target variable
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pc_df['Diagnosis'] = y

# Define colors for plotting
colors = {'B': 'blue', 'M': 'red'}

# Plot the scatter plot
plt.figure(figsize=(10, 8))
plt.scatter(pc_df['PC1'], pc_df['PC2'], c=[colors[i] for i in pc_df['Diagnosis']], alpha=0.6)
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('Scatter Plot of First Two Principal Components')
plt.show()

```

```python
# Assuming data2 is already defined
X = data2[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA with three components
pca = PCA(n_components=3)
principal_components = pca.fit_transform(X_scaled)

# Create a DataFrame with the principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])

# Plot the PCA results in a scatter plot with varied colors
plt.figure(figsize=(12, 6))

# Scatter plot for the first two principal components with varied colors
plt.subplot(1, 2, 1)
plt.scatter(pc_df['PC1'], pc_df['PC2'], c=pc_df['PC1'], cmap='viridis', marker='o')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.title('PCA: PC1 vs PC2')

# Scatter plot for the second and third principal components with varied colors
plt.subplot(1, 2, 2)
plt.scatter(pc_df['PC2'], pc_df['PC3'], c=pc_df['PC2'], cmap='viridis', marker='o')
plt.xlabel('Principal Component 2 (PC2)')
plt.ylabel('Principal Component 3 (PC3)')
plt.title('PCA: PC2 vs PC3')

plt.tight_layout()
plt.show()
```

#this plot below provides  insights into how each original feature contributes to the top three principal components. Positive and negative values indicate the direction of the contribution, and the length of the bars represents the magnitude. This visualization helps in understanding which features are more influential in each principal component and provides a compact representation of the data in a reduced-dimensional space.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Select more features for PCA
features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']

# Add more features to the features list
X = data2[features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA with three components
pca = PCA(n_components=3)
principal_components = pca.fit_transform(X_scaled)

# Create a DataFrame with the principal components
pc_df = pd.DataFrame(data=pca.components_.T, columns=[f'PC{i+1}' for i in range(3)], index=features)

# Transpose the DataFrame so that features become legend entries
pc_df_T = pc_df.T

# Plot the top three principal components with a smaller figure size
plt.figure(figsize=(10, 6))  # Adjust the figsize to make the figure smaller

# Plot the bar chart
pc_df_T.plot(kind='bar', rot=0, width=0.7, colormap='viridis')

# Add title and labels
plt.title('Top Three Principal Components vs Original Features', fontsize=16)
plt.xlabel('Principal Components', fontsize=14)
plt.ylabel('Component Value', fontsize=14)

# Add a legend outside the plot for better visibility
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Original Features', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()





```

#the code performs hierarchical clustering on the specified features, standardizes the data, and then visualizes the hierarchical structure of the clusters using a dendrogram. The dendrogram illustrates the relationships and distances between data points, allowing for an interpretation of the clustering structure.

```python
from scipy.cluster import hierarchy

# Select the features for clustering
X = data2[['radius_mean', 'texture_mean', 'perimeter_mean']]  \

# Standardize the features
scaler = StandardScaler()
std_X = scaler.fit_transform(X)

# Perform hierarchical clustering
linkage_matrix = hierarchy.linkage(std_X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 6))
dendrogram = hierarchy.dendrogram(linkage_matrix, truncate_mode='lastp', p=40)

# Customize the plot
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

In summary, this code combines PCA with hierarchical clustering, assigns cluster labels to the data, and then visualizes the mean values of features for each cluster using a heatmap. The provided interpretation for Cluster 1 highlights specific characteristics observed in the mean values of certain features.

```python


# Select the features for PCA
X = data2[['radius_mean', 'texture_mean', 'perimeter_mean']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA with three components
pca = PCA(n_components=3)
principal_components = pca.fit_transform(X_scaled)

# Perform hierarchical clustering on PCA components
linkage_matrix = hierarchy.linkage(principal_components, method='ward')

# Get cluster labels
num_clusters = 3  # Choose the number of clusters
cluster_labels = hierarchy.fcluster(linkage_matrix, num_clusters, criterion='maxclust')

# Add cluster labels to the original data
data_with_labels = pd.concat([data2, pd.DataFrame({'Cluster': cluster_labels})], axis=1)

# Create a heatmap of the cluster labels
plt.figure(figsize=(10, 8))
sns.heatmap(data_with_labels.groupby('Cluster').mean().transpose(), annot=True, cmap='viridis', fmt='.2f', linewidths=0.5)
plt.title('PCA Clusters - Heatmap')
plt.show()
```
